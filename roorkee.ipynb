```python

from pdf2image import convert_from_path

# images = convert_from_path("/content/new.jpeg")
# images[0]

from PIL import Image

images = [Image.open(f"/content/bday.png")]
```

```python

from byaldi import RAGMultiModalModel

RAG = RAGMultiModalModel.from_pretrained("vidore/colpali")
```

```python
RAG.index(
    input_path="/content/bday.png",
    index_name="image_index", # index will be saved at index_root/index_name/
    store_collection_with_index=False,
    overwrite=True
)
```

```python
text_query = "Explain the image"
results = RAG.search(text_query, k=1)
results
```

```python
import torch
from transformers import Qwen2VLForConditionalGeneration, AutoTokenizer, AutoProcessor
from qwen_vl_utils import process_vision_info

model = Qwen2VLForConditionalGeneration.from_pretrained("Qwen/Qwen2-VL-2B-Instruct",
                                                        trust_remote_code=True, torch_dtype=torch.bfloat16).cuda().eval()
```

```python

processor = AutoProcessor.from_pretrained("Qwen/Qwen2-VL-2B-Instruct", trust_remote_code=True)

image_index = results[0]["page_num"] - 1
messages = [
    {
        "role": "user",
        "content": [
            {
                "type": "image",
                "image": images[image_index],
            },
            {"type": "text", "text": text_query},
        ],
    }
]
```

```python

text = processor.apply_chat_template(
    messages, tokenize=False, add_generation_prompt=True
)

image_inputs, _ = process_vision_info(messages)


# Assuming you are using something like PIL for images, resize if necessary
image_inputs = [img.resize((256, 256)) for img in image_inputs]  # Example size
```

```python
from torch.cuda.amp import autocast

with autocast():
    inputs = processor(
        text=[text],
        images=image_inputs,
        padding=True,
        return_tensors="pt",
    )

batch_size = 8  # adjust as necessary
for i in range(0, len(image_inputs), batch_size):
    batch_image_inputs = image_inputs[i:i + batch_size]
    inputs = processor(
        text=[text],
        images=batch_image_inputs,
        padding=True,
        return_tensors="pt",
    )
    inputs = inputs.to("cuda")
    # process batch


# Clear CUDA memory before processing
torch.cuda.empty_cache()

inputs = processor(
    text=[text],
    images=image_inputs,
    padding=True,
    return_tensors="pt",
)
inputs = inputs.to("cuda")

# After processing, free up memory
torch.cuda.empty_cache()
```

```python
generated_ids = model.generate(**inputs, max_new_tokens=50)

generated_ids
```

```python
generated_ids_trimmed = [
    out_ids[len(in_ids) :] for in_ids, out_ids in zip(inputs.input_ids, generated_ids)
]
output_text = processor.batch_decode(
    generated_ids_trimmed, skip_special_tokens=True, clean_up_tokenization_spaces=False
)
```
